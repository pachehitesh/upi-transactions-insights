## **2. Handle Outliers (IQR Method)**

```python
# Define numeric columns to check for outliers
numeric_cols = ['Transaction (thousands)', 'Balance (thousands)', 'Age']

# Detect and optionally remove outliers using IQR
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Outliers detected in {col}: {outliers.shape[0]}")
```

```python
# Convert Time to datetime.time if not already
df['Corrected Time'] = pd.to_datetime(df['Time'], errors='coerce').dt.time

# Extract hour for analysis
df['Hour'] = pd.to_datetime(df['Time'], errors='coerce').dt.hour
```

## **Deep Insight Extraction**
 **1. Segmented Transaction Behavior by Age Group**

```python
# Average transaction per age group
age_group_summary = df.groupby('Age Groups')['Transaction (thousands)'].agg(['mean', 'count']).reset_index()
age_group_summary
```

**Insight:**

Younger users (e.g., Youth) make frequent but smaller transactions.

Senior Adults tend to make fewer but higher-value transactions.



**2.Transaction Time  Trends**

```python
hourly_trend = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x=hourly_trend.index, y=hourly_trend.values, marker='o')
plt.title("Transaction Volume by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Transactions")
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

**Insight:**

Peak UPI transaction activity occurs between **11 AM to 2 PM** and **6 PM to 8 PM.**

Very low activity occurs **late night (12 AM – 5 AM).**


**3. Top Performing Banks in Transactions**

```python
top_senders = df['Sent From Bank'].value_counts().head(5)
top_receivers = df['Bank Received'].value_counts().head(5)

top_senders_df = top_senders.reset_index().rename(columns={'index': 'Bank', 'Sent From Bank': 'Send Count'})
top_receivers_df = top_receivers.reset_index().rename(columns={'index': 'Bank', 'Bank Received': 'Receive Count'})
```

**Insight:**

A few banks dominate both sending and receiving ends.

**Strategic partnerships** with top banks can boost transaction volume.

**4. Gender Differences in Transaction Patterns**

```python
gender_summary=df.groupby('Gender')['Transaction (thousands)'].agg(['mean','count'])
gender_summary
```
